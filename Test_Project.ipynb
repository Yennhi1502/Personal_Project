```python
import re

def quality_check(password):
    if not 8 <= len(password) <= 15:
        return password,"Khong phai la password"
    if not re.search(r'[!@#$^]', password):
        return password,"Khong phai la password"
    if not re.search(r'\d', password):
        return password,"Khong phai la password"
    if not re.search(r'[A-Z]', password):
        return password,"Khong phai la password"
    if not re.search(r'[a-z]', password):
        return password,"Khong phai la password"
    return password,"La password"
password = input("Nhập password: ")
quality_check(password)
```

```python
# Câu 3:  Bạn hãy xây dựng một hàm nhận tham số truyền vào là một danh sách bất kỳ chỉ chứa các số. 
# Hàm tính toán và  trả về số lớn nhì trong danh sách. Với trường hợp danh sách gồm các số bằng nhau, hàm trả về số lớn nhất. Nếu có nhiều giá trị bằng số lớn nhì trong danh sách, hàm trả về 1 giá trị duy nhất. (1.5đ)
# Ví dụ: - Với danh sách [1,1,1,1,1,1] -> Hàm trả về 1
# Với danh sách [1,2,1,1,2,3] -> Hàm trả về 2  
```

```python
num_list = int(input("Nhập độ dài danh sách: "))
lst = [] 
count = 0
while count < num_list: 
    number = int(input(f"Nhập số vị trí {count}: ")) 
    lst.append(number)
    count += 1 
lst.sort() 
res = []

for i in lst:
    if i not in res:
        res.append(i)

num_second = res[len(res) - 2]
print("Số lớn thứ 2 trong mảng là: ", num_second)
```
